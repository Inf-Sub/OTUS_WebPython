```python
soup = bs(response.text, "html.parser")
print(soup)
```

```python
soup.title.text
```

```python
soup.h1
```

```python
soup.h1.text
```

```python
soup.h1["class"]
```

```python
div_tile = soup.find("div", {"class": "blog__tile blog__tile_slim"})
print(div_tile)
```

```python
div_tile["class"]
```

```python
headers_2 = soup.find_all("h2")
print(headers_2)
```

```python
for header2 in headers_2:
    print(header2.text)
```

```python
h1s_and_h2s = soup.find_all(["h1", "h2"])
print(h1s_and_h2s)
```

```python
for header in h1s_and_h2s:
    print(header.text)
```

```python
list(vars(header).keys())
```

```python
div_tile.name
```

```python
for header in h1s_and_h2s:
    print(header.name, header.text)
```

```python
similar_posts_selector = "body > div.body-wrapper > div > div.blog.js-blog > div.blog__content > div > div > div.container__col.container__col_8.container__col_md-12 > div.blog-tile-wrapper"
similar_posts = soup.select_one(similar_posts_selector)
print(similar_posts)
```

```python
similar_posts.name
```

```python
similar_posts_links = similar_posts.find_all("a", {"class": "blog-tile__item-title"})
print(similar_posts_links)
```

```python
for link in similar_posts_links:
    print("*", link["title"], link["href"])
```
